```python
# importing Pandas library 
import pandas as pd
```

```python
# reading csv file for first 6 months of 2020

jan_2020 = pd.read_csv("202001-citibike-tripdata.csv")
feb_2020 = pd.read_csv("202002-citibike-tripdata.csv")
mar_2020 = pd.read_csv("202003-citibike-tripdata.csv")
apr_2020 = pd.read_csv("202004-citibike-tripdata.csv")
may_2020 = pd.read_csv("202005-citibike-tripdata.csv")
jun_2020 = pd.read_csv("202006-citibike-tripdata.csv")
```

```python
# using a .concat for union of dataframe
city_bike_2020 = pd.concat([jan_2020, feb_2020, mar_2020, apr_2020, may_2020, jun_2020])

# Printing the dataframe 
city_bike_2020
```

```python
# Replace "0" with "Unknown"; "1" with "Male", "2" with "Female"
city_bike_2020['gender'] = city_bike_2020['gender'].replace([0,1,2],["Unknown", "Male", "Female"])

# Printing the dataframe 
city_bike_2020
```

```python
# Checking to see count of datapoints
city_bike_2020.count()
```

```python
# Dropping NaN
city_bike_2020 = city_bike_2020.dropna()
city_bike_2020.count()
```

```python
# return the shape 
city_bike_2020.shape
```

```python
# calling describe method 
city_bike_2020.describe()
```

```python
# return the dtype of each column 
city_bike_2020.dtypes
```

```python
# Save DataFrame to CSV
city_bike_2020.to_csv("city_bike_2020.csv", index=False, header=True)
```
